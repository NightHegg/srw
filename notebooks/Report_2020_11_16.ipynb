## Решение задач с применением метода Шварца:

В нашей задаче мы проводим декомпозицию области и представляем её (область) в виде объединения конечного числа пересекающихся подобластей $\Omega_1, ..., \Omega_N$, где N - число подобластей. 
- $\delta \Omega_D$ - участок границы, на котором заданы кинематические условия; 
- $\delta \Omega_N$ - участок границы, на котором заданы силовые условия; 
- количество итераций для понимания поведения графиков - 5; 

Рассмотрим несколько случаев проставления условий Шварца на границе:

#### Первый случай:
- для $\Omega_1$: 
\begin{aligned}
Lu_1^{n+1} = f_1, u_1 \in \Omega_1\\
u_1^{n+1} = u^n, u_1 \in \delta\Omega_1 \cap \Omega_2
\end{aligned}
Учтём с этими условиями, что ($Г_1 = \delta\Omega_1 \cap \Omega_2$):
 - $Г_1 \cap \delta\Omega_D = 0$;
 - $Г_1 \cap \delta\Omega_N = 0$;
- для $\Omega_2$: 
\begin{aligned}
Lu_2^{n+1} = f_2, u_2 \in \Omega_2\\
u_2^{n+1} = u^n, u_2 \in \delta\Omega_2 \cap \Omega_1
\end{aligned}
